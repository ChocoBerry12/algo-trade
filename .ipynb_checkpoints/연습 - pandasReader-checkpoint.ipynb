> 이미 여러분은 국내 증권사의 API를 이용해 주식 데이터를 받아오는 법을 배웠습니다. 실시간 데이터라면 증권사의 API를 사용하는 것이 편리하지만 일봉 데이터 정도는 데이터의 양이 많지 않기 때문에 야후(www.yahoo.com)나 구글 (www.google.com) 같은 인터넷 웹사이트를 통해 구할 수도 있습니다.

* web 에서 data 를 dataFrame 형태로 변환해줌

```python
%matplotlib inline
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2018, 1, 16)
end = datetime.datetime(2019, 6, 20)
gs = web.DataReader("005930.KS", "yahoo", start, end) # web 데이터를 dataFrame 으로 변환
gs
```

```python
print(gs.info(), '\n') # 객체정보 요약
gs['High'].plot()
```
